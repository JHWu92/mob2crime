```python
import geopandas as gp
```

```python
cities = gp.read_file('data/cities_mexico.geojson')
```

```python
vor = gp.read_file('data/mex_vor_08.geojson')
```

```python
%matplotlib inline
```

```python
vor_inter_cities = gp.sjoin(vor, cities)
```

```python
vor_inter_cities
```

```python
vor_clip = []
for gtid, cname in vor_inter_cities[['gtid', 'cname']].values:
    vor_clip.append((gtid, cname,vor[vor.gtid==gtid].geometry.intersection(cities[cities.cname==cname].geometry)))
```

```python
vor_clip = []
for i,ir,gtid, cname in vor_inter_cities[['index', 'index_right','gtid', 'cname']].values:
#     if len(vor_clip)%30==0:
#         print(len(vor_clip))
    vor_clip.append((gtid, cname,vor.loc[i].geometry.intersection(cities.loc[ir].geometry)))
```

```python
vor_clip = gp.GeoDataFrame(vor_clip,columns=['gtid', 'cname','geometry'])
```

```python
vor_clip.to_file('data/mex_vor_08_clip.geojson', driver='GeoJSON')
```
